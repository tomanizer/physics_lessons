## Measurement

### The Metric System (International System of Units)

https://www.nist.gov/content/si-redefinition


#### Base Units


| Name     | Symbol | Measures what             |
|----------|--------|---------------------------|
| second   | $s$      | **time**                      |
| meter    | $m$      | **length**                    |
| kilogram | $kg$     | **mass**                      |
| ampere   | $A$      | **electric current**          |
| kelvin   | $K$      | **thermodynamic temperature** |
| mole     | $mol$    | **amount of substance**       |
| candela  | $cd$     | **luminous intensity**        |


#### Metric Prefixes

For larger units

| Name  | Symbol | Value | Value name |
|-------|--------|-------|-----------|
| giga  | $G$      | $10^9$  | **Billion**   |
| mega  | $M$      | $10^6$  | **Million**   |
| kilo  | $k$      | $10^3$  | **Thousand**  |
| hecto | $h$      | $10^2$  | **Hundred**   |
| deka  | $da$     | $10^1$  | **Ten**       |

For smaller units

| Name  | Symbol | Value | Value name  |
|-------|--------|-------|------------|
| deci  | $d$      | $10^{-1}$ | **Tenth**      |
| centi | $c$      | $10^{-2}$ | **Hundredth**  |
| milli | $m$      | $10^{-3}$ | **Thousandth** |
| micro | $\mu$      | $10^{-6}$ | **Millionth**  |
| nano  | $n$      | $10^{-9}$ | **Billionth**  |


#### Derived Units

| Coherent derived units   | Symbol           | Measures what                     |
|--------------------------|------------------|-----------------------------------|
| square meter             | $m^2$            | **area**                              |
| cubic meter              | $m^3$            | **volume**                            |
| meter/second             | $m/s$            | **speed, velocity**                   |
| meter per second per second | $\frac{m}{s^2}$          | **acceleration**                      |
| reciprocal meter         | $\frac{1}{m}$     | **wave number**                       |
| kilogram per cubic meter | $\frac{kg}{m^3}$ | **mass density**                      |
| cubic meter per kilogram | $\frac{m^3}{kg}$ | **specific volume**                   |
| ampere per square meter  | $\frac{A}{m^2}$  | **current density**                   |
| ampere per meter         | $\frac{A}{m}$     | **magnetic field strength**           |
| mole per cubic meter     | $\frac{mol}{m^3}$ | **amount of substance concentration** |
| candela per square meter | $\frac{cd}{m^2}$       | **luminance**                         |

#### Derived Quantities

| Derived Quantity   unit | Symbol | Derived units | Base Units    | Measuring what                     |
|-------------------------|--------|---------------|---------------|------------------------------------|
| degree   Celsius        | $°C$     |              | $K$             | **Celsius temperature**                |
| newton                  | $N$      |              | $kg \frac{m}{s^2}$      | **force**                              |
| joule                   | $J$      | $N·m$         | $kg \frac{m^2}{s^2}$     | **energy, work, quantity of heat**   |
| watt                    | $W$      | $\frac{J}{s}$ | $kg \frac{m^2}{s^3}$     | **power**                              |
| volt                    | $V$      | $\frac{W}{A}$ |  $kg \frac{m^2}{s^3 A}$ | **electric potential difference**     |
| ohm                    | $\Omega$ | $\frac{V}{A}$ |  $kg \frac{m^2}{s^3·A^2}$ | **electric resistance**                  | 
| pascal                  | $Pa$     | $\frac{N}{m^2}$ | $\frac{kg}{m·s^2}$    | **pressure, stress**                   |
| hertz                   | $Hz$     |              | $\frac{1}{s}$           | **frequency**                          |
|becquerel                |$Bq$      |              |$\frac{1}{s}$            | **activity (of a radionuclide)**       |
